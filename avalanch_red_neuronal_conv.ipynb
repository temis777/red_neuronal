normalizar la imagen entre0y1

```python
def read(filename):
    
    im = np.array(PIL.Image.open(filename))
    if im.dtype == np.uint8: im = im / (2**8-1)
    elif im.dtype == np.uint16: im = im / (2**16-1)
    elif im.dtype == np.uint32: im = im / (2**32-1)
    return im
```

```python
input_array = read(input_filenames[1])
print('Shape of the first image:', input_array.shape)
print('DType of the first image:', input_array.dtype)

plt.figure()
plt.imshow(input_array, vmin=0, vmax=1, cmap="gray")
plt.colorbar()
plt.show()
plt.close()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import glob

import PIL.Image
```

```python
test_water = '/content/train/input'
input=[]
image_size=150
```

```python
for img in os.listdir(test_water):
  img = cv2.imread(os.path.join(test_water,img))
  img_resize = cv2.resize(img,(img_size,img_size))
  input.append(img_resize)
```

```python
print(len(input))
print(input)
```

```python
test_water = np.array(input)
print(test_water.shape)
```

```python
print(input[2].shape)
plt.figure()
plt.imshow(np.squeeze(input[2]))
plt.colorbar()
plt.grid(False)
plt.show()
```

JUNTAR IMAGENES

```python
images = np.concatenate([test_water,input])
#sirve para ver cuantas imagenes tenemos
print(len(images))
# lista de imagenes ya en arreglo
Images = np.array(images)
print(Images.shape)
#numero de imagenes, tamao de imagen, tamao de image, canales de colores
```

```python
etiquetas_test_water = np.repeat(1,1000)
print(len(etiquetas_test_water))
print(etiquetas_test_water)
```

CLASES


```python
#NOMBRE DE LAS CLASES PARA 0 = TEST Y PARA 1 = WATER
class_names=['input','test_water']
```

JUNTAR ETIQUETAS EN UNA SOLA

```python
# SEUNEN LAS ETIQUETAS ATRAVES DE LA FUNCION NP.CPNCATENATE
labels = np.concatenate([etiquetas_test_water,etiquetas_input])
#ver cuantas etiquetas tenemos
print(len(labels))
print(labels)
#meterlas en un arreglo y mostrar
Labels =np.array(labels)
print(Labels.shape)
```
